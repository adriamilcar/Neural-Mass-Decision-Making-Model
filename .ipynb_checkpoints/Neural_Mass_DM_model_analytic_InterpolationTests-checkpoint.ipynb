```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad
from scipy.integrate import odeint
from scipy.optimize import fsolve
from scipy.interpolate import Rbf
from scipy.spatial.distance import cdist
```

## System equations

```python
def logistic_function(U, F_max, x_0, k):
    exponent = -k * (U - x_0)
    with np.errstate(over='ignore'):
        exp_term = np.exp(exponent) if exponent > -700 else 0.0   # Threshold to prevent overflow, typically -700 for double precision
    return F_max / (1 + exp_term)


def logistic_derivative(U, F_max, x_0, k):
    f_U = logistic_function(U, F_max, x_0, k)
    return f_U * (1 - f_U / F_max)


def jacobian(U1, U2, w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau):
    H1 = w1 * U1 + I1 - Q * w2 * U2 - (1 - Q) * w3 * logistic_function(U1 + U2, F_max, x_0, k)
    H2 = U1 + U2
    H3 = w1 * U2 + I2 - Q * w2 * U1 - (1 - Q) * w3 * logistic_function(U1 + U2, F_max, x_0, k)

    f_prime_H1 = logistic_derivative(H1, F_max, x_0, k)
    f_prime_H2 = logistic_derivative(H2, F_max, x_0, k)
    f_prime_H3 = logistic_derivative(H3, F_max, x_0, k)

    J11 = (-1 + f_prime_H1 * (w1 - (1 - Q) * w3 * f_prime_H2)) / tau
    J12 = (-f_prime_H1 * (Q * w2 - (1 - Q) * w3 * f_prime_H2)) / tau
    J21 = (-f_prime_H3 * (Q * w2 - (1 - Q) * w3 * f_prime_H2)) / tau
    J22 = (-1 + f_prime_H3 * (w1 - (1 - Q) * w3 * f_prime_H2)) / tau

    J = np.array([[J11, J12],
                  [J21, J22]])
    return J


def system_dynamics(U, w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau):
    U1, U2 = U
    eq1 = (-U1 + logistic_function(w1 * U1 + I1 - Q * w2 * U2 - (1 - Q) * w3 * logistic_function(U1 + U2, F_max, x_0, k), F_max, x_0, k)) / tau
    eq2 = (-U2 + logistic_function(w1 * U2 + I2 - Q * w2 * U1 - (1 - Q) * w3 * logistic_function(U1 + U2, F_max, x_0, k), F_max, x_0, k)) / tau
    return [eq1, eq2]
```

## Params

```python
# Params
w1, w2, w3 = 5.0, 4.0, 4.0  # Parameters for the logistic function
Q = 0.75  # Trade-off parameter
I1, I2 = 4., 4.  # Input currents
F_max, x_0, k = 40.0, 15.0, 1./22.  # Parameters for the logistic function
tau = 20e-3  # Time constant
```

## Fixed point analysis

```python
def classify_fixed_point(J):
    eigenvalues = np.linalg.eigvals(J)
    if np.all(eigenvalues < 0):
        return 'Attractor'
    elif np.all(eigenvalues > 0):
        return 'Repeller'
    else:
        return 'Saddle'
    
    
def closest_fixed_point(U1, U2, fixed_points):
    """
    Finds the closest fixed point to the given point U.

    Parameters:
    U: tuple
        A point in the space (U1, U2)
    fixed_points: list of tuples
        A list where each element is a tuple containing the fixed point coordinates and its type.

    Returns:
    A tuple representing the closest fixed point and its type.
    """
    closest_point = None
    min_distance = float('inf')  # Start with infinity as the minimum distance

    for fp, fp_type in fixed_points:
        distance = np.sqrt((U1 - fp[0])**2 + (U2 - fp[1])**2)

        if distance < min_distance:
            min_distance = distance
            closest_point = fp

    return closest_point


def fixed_point_dist_weights(U1, U2, fixed_points):
    distances = np.array([np.sqrt((U1 - fp[0])**2 + (U2 - fp[1])**2) for fp in fixed_points])
    with np.errstate(divide='ignore', invalid='ignore'):
        distances = np.where(distances == 0, np.finfo(float).eps, distances)
        weights = 1/distances
        weights /= np.sum(weights)  # normalize weights
    return weights


def rbf_weights(U1, U2, fixed_points, function='multiquadric', epsilon=1):
    """
    Computes the RBF weights for a given point with respect to the fixed points.

    Parameters:
    fixed_points (array): An array of fixed points in the space.
    function (str): Type of RBF function to use.
    epsilon (float): A parameter that defines the smoothness of the interpolation.

    Returns:
    array: The weights for each fixed point.
    """
    point = np.array([U1, U2])[np.newaxis, :]
    distances = cdist(fixed_points, point, 'euclidean').flatten()

    if function == 'multiquadric':
        weights = np.sqrt((1.0/epsilon * distances)**2 + 1)
    elif function == 'inverse':
        weights = 1.0 / np.sqrt((1.0/epsilon * distances)**2 + 1)
    elif function == 'gaussian':
        weights = np.exp(-(1.0/epsilon * distances)**2)

    weights /= np.sum(weights)

    return weights

def shepards_weights(U1, U2, fixed_points, power_parameter=2):
    """
    Computes the weights for a given point using Shepard's method.

    Parameters:
    fixed_points (array): An array of fixed points in the space.
    power_parameter (float): The power parameter that determines the rate of falloff.

    Returns:
    array: The weights for each fixed point.
    """
    point = np.array([U1, U2])[np.newaxis, :]
    distances = cdist(fixed_points, point, 'euclidean').flatten()
    with np.errstate(divide='ignore', invalid='ignore'):
        distances_with_epsilon = np.where(distances == 0, np.finfo(float).eps, distances)
        weights = (1/distances_with_epsilon)**power_parameter
        weights /= np.sum(weights)
    return weights


def find_fixed_points_and_classify(w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau):
    def func(U):
        return system_dynamics(U, w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau)
    
    initial_guesses = [(i, j) for i in np.linspace(0, F_max, 40) for j in np.linspace(0, F_max, 40)]
    fixed_points = []
    fixed_points_info = []
    for initial_guess in initial_guesses:
        solution, infodict, ier, mesg = fsolve(func, initial_guess, full_output=True)
        if ier == 1 and not np.any([np.allclose(solution, fp) for fp in fixed_points]):
            J = jacobian(solution[0], solution[1], w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau)
            classification = classify_fixed_point(J)
            fixed_points.append(solution)
            fixed_points_info.append(classification)
    
    return fixed_points, fixed_points_info
```

```python
fixed_points, fixed_points_info = find_fixed_points_and_classify(w1, w2, w3, Q, I1, I2, F_max, x_0, k, tau)
for point, classification in list(zip(fixed_points, fixed_points_info)):
    print(f"Fixed Point: {point}, Classification: {classification}")
```

## Energy landscape

```python
# Define the differential equations system
def dU_dt(U1, U2, w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau):
    fU1 = logistic_function(w1*U1 + I1 - Q*w2*U2 - (1-Q)*w3*logistic_function(U1 + U2, F_max, x_0, k), F_max, x_0, k)
    fU2 = logistic_function(w1*U2 + I2 - Q*w2*U1 - (1-Q)*w3*logistic_function(U1 + U2, F_max, x_0, k), F_max, x_0, k)
    dU1dt = (-U1 + fU1) / tau
    dU2dt = (-U2 + fU2) / tau
    return dU1dt, dU2dt

def potential(U1, U2, w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau, ref_potential=[0,0]):
    # Define the integrands for U1 and U2
        def integrand_U1(x):
            return -dU_dt(x, U2, w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau)[0]

        def integrand_U2(y):
            return -dU_dt(U1, y, w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau)[1]

        # Perform the numerical integration
        integral_U1, _ = quad(integrand_U1, ref_potential[0], U1)
        integral_U2, _ = quad(integrand_U2, ref_potential[1], U2)

        # The potential at this point is the sum of the two integrals
        V = integral_U1 + integral_U2
        return V


# Numerically estimate the potential
def potential_surface(w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau, fixed_points, resolution=100):
    U1_range = np.linspace(0, F_max, resolution)
    U2_range = np.linspace(0, F_max, resolution)
    V = np.zeros((resolution, resolution))
    
    for i, U1 in enumerate(U1_range):
        for j, U2 in enumerate(U2_range):
            weights = rbf_weights(U1, U2, fixed_points, function='inverse')
            weighted_v = 0.
            for k, fp in enumerate(fixed_points):
                v = potential(U1, U2, w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau, fp)
                weighted_v += weights[k] * v
            V[i, j] = weighted_v
    
    V -= np.min(V)
    V /= np.max(V)
    
    U1_mesh, U2_mesh = np.meshgrid(U1_range, U2_range)
    
    return U1_mesh, U2_mesh, V


def plot_potential_surface(U1, U2, V, fixed_points, fixed_points_info):
    fig = plt.figure(figsize=(10, 7), dpi=600)
    ax = fig.add_subplot(111, projection='3d')
    
    surf = ax.plot_surface(U1, U2, V, cmap='viridis', alpha=0.8)
    
    contour = ax.contour(U1, U2, V, 50, cmap='viridis', linewidths=.5, offset=np.min(V))
    
    point_offset = 0.02
    for point, ptype in list(zip(fixed_points, fixed_points_info)):
        indx_U1 = int(point[0]/F_max*U1.shape[0]) 
        indx_U2 = int(point[1]/F_max*U2.shape[0]) 
        if ptype == "Attractor":
            ax.scatter(point[0], point[1], V[indx_U1, indx_U2]+point_offset, color='g', s=100)
        elif ptype == "Saddle":
            ax.scatter(point[0], point[1], V[indx_U1, indx_U2]+point_offset, color='r', s=100)
        else:
            ax.scatter(point[0], point[1], V[indx_U1, indx_U2]+point_offset, color='black', s=100)
    
    ax.view_init(elev=25, azim=-105)
    
    fig.colorbar(surf, shrink=0.4, aspect=20)
    ax.set_xlabel('U1')
    #ax.invert_xaxis()
    ax.set_ylabel('U2')
    ax.set_zlabel('V')
    
    plt.tight_layout()
    plt.show()
```

```python
# Estimate potential
U1_mesh, U2_mesh, V = potential_surface(w1, w2, w3, I1, I2, Q, F_max, x_0, k, tau, fixed_points, resolution=50)
```

```python
# Plot the potential
plot_potential_surface(U1_mesh, U2_mesh, V, fixed_points, fixed_points_info)
```
